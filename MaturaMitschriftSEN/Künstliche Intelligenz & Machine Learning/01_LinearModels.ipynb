# Linear Models:

```python
import matplotlib.pyplot as plt

import numpy as np
np.random.seed(42)
```

```python
m = 100
X = np.random.rand(m, 1) * 2

#f(x) = 4 + 3 * x
y = 4 + 3 * X + np.random.randn(m, 1)
```

Mit .randn wird ein Rauschen simuliert

```python
plt.figure(figsize=(6, 4))
plt.plot(X, y, "b.")
plt.grid()
plt.show()
```

d = 4 ....theta0 = 4
k = 3 ....theta = 3

```python
X_b = np.c_[np.ones((100, 1)), X]
```

Wir verwenden Lineare Regression (aus dem np Paket)

```python
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

```python
X_p = 2 * np.random.rand(m, 1)
y_p = theta_best[0] + theta_best[1] * X_p

plt.plot(X_p, y_p, "r-")
plt.plot(X, y, "b.")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

.intercept ist das Theta0 / .coef ist das Theta1,2,3,...

```python
lin_reg.intercept_, lin_reg.coef_
```

0.5 * x^2 + x + 2

```python
m = 100
X = 6 * np.random.rand(m , 1) - 3

y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)
```

```python
plt.figure(figsize=(6, 4))
plt.plot(X, y, "b.")
plt.grid()
plt.show()
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X,y)
```

```python
print(lin_reg.intercept_[0])
print(lin_reg.coef_[0,0])
```

```python
X_p = 6 * np.random.rand(m, 1) - 3
y_p = theta_best[0] + theta_best[0,0] * X_p

plt.plot(X_p, y_p, "r-")
plt.plot(X, y, "b.")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)
```

Eine Zweite Spalte des ersten Wertes zum Quadrat hinzufügen

```python
X_poly
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
```

```python
print(lin_reg.intercept_[0])
print(lin_reg.coef_[0,0])
print(lin_reg.coef_[0,1])
```

```python
y_p = lin_reg.intercept_[0] + lin_reg.coef_[0,0] * X + lin_reg.coef_[0,1] * X**2
```

```python
plt.plot(X_p, y_p, "r.")
plt.plot(X, y, "b.")
plt.show()
```
